Customized version: bstats

```python
def bstats(vec):
    s = {'num_element':len(vec), 'minimo':vec.min(),
         'maximo':vec.max(), 'media':vec.mean(),
         'desvio_padrao':vec.std()}
    return s
```

```python
print(bstats(x))
print(bstats(y))
print(bstats(z[:,1]))
```

Kernel Density Estimates

http://www.scipy.org/doc/api_docs/SciPy.stats.kde.gaussian_kde.html

```python
nbins=100
plt.hist(y, bins=nbins, normed=True, align='mid')
plt.title('Server Response')
plt.ylabel('time (s)')
```

```python
plt.figure(2)    
gkde = gaussian_kde(y)
plt.plot(arange(0,(1.01*(max(y)-min(y))),.1), gkde.evaluate(arange(0,(1.01*(max(y)-min(y))),.1)))
plt.title('Server Response')
plt.ylabel('time (s)')
plt.show()
```

```python
plt.subplot(2, 1, 1)
nbins=100
plt.hist(y, bins=nbins, normed=True, align='mid')
plt.title('Server Response')
plt.ylabel('time (s)')

plt.subplot(2, 1, 2)
gkde = gaussian_kde(y)
plt.plot(arange(0,(1.01*(max(y)-min(y))),.1), gkde.evaluate(arange(0,(1.01*(max(y)-min(y))),.1)))   
plt.xlabel('Responses with a given time')
plt.ylabel('time (s)')

plt.show()
```
